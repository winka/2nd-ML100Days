```python
import os 
import pandas as pd
import numpy as np 


dir_data='./data/'

train_file=os.path.join(dir_data,'titanic_train.csv')
train=pd.read_csv(train_file)

test_file=os.path.join(dir_data,'titanic_test.csv')
test=pd.read_csv(test_file)
```

```python
train=train.drop(['PassengerId', 'Survived'],axis=1)#在train的資料中移除passangerid , survived
test=test.drop(['PassengerId'],axis=1)
df=pd.concat([train,test]) #合併兩個資料
df.head()
```

```python
int_features=[]
float_features=[]
object_features=[]

for d_type,feature in zip(df.dtypes,df.columns):
    if d_type=='int':
        int_features.append(feature)
    if d_type=='float':
        float_features.append(feature)
    if d_type=='object':
        object_features.append(feature)

print(f'{len(int_features)} Integer Features : {int_features}\n')
print('Interger mean max nunique:')
print(df[int_features].mean())
print(df[int_features].max())
print(df[int_features].nunique())
print('------------------------------------------------------------')
print(f'{len(float_features)} Float Features : {float_features}\n')
print('Float mean max nunique:')
print(df[float_features].mean())
print(df[float_features].max())
print(df[float_features].nunique())
print('------------------------------------------------------------')
print(f'{len(object_features)} Object Features : {object_features}')
print('Object mean max nunique:')
print(df[object_features].mean())
print(df[object_features].max())
print(df[object_features].nunique())
```
