```python
import os 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
import warnings
warnings.filterwarnings('ignore')
l=[]
for i in range(100):
    a=round(np.random.uniform(-1, 1),2)
    l+=[a]
marray=pd.DataFrame(l)
print(marray)


plt.figure(figsize = (10, 12))
sns.heatmap(marray[[0]], cmap = plt.cm.RdYlBu_r, vmin = -1, annot = True, vmax = 1)#heapmap使用的資料為2維 所以先將資料一維轉2維
plt.title('Random Heatmap');
```

```python
arr1=[]
arr2=[]
arr3=[]
for i in range(1000):
    a=round(np.random.uniform(-1,1),2)
    b=round(np.random.uniform(-1,1),2)
    c=np.random.uniform(-1,1)
    arr1+=[a]
    arr2+=[b]
    arr3+=[c]

nrow=1000
ncol=3
dict = {"array1": arr1,  
        "array2": arr2,
        "array3": arr3
       }
select_df = pd.DataFrame(dict)

grid = sns.PairGrid(data = select_df, size = 3, diag_sharey=False, 
                     vars = select_df)

grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(sns.kdeplot) 
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.show()
```
