```python
import os 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件


dir_data='./data/'

train_file=os.path.join(dir_data,'application_train.csv')
train=pd.read_csv(train_file)
train.head()
```

```python
unique_house_type = train.HOUSETYPE_MODE.unique()

nrows = len(unique_house_type)
ncols = nrows // 2

plt.figure(figsize=(10,30))

for i in range(len(unique_house_type)):
    plt.subplot(nrows,ncols,i+1)
    
    train.loc[(train.HOUSETYPE_MODE==unique_house_type[i]) ,'AMT_ANNUITY'].hist()
    plt.title(str(unique_house_type[i]))
plt.show()    
```
