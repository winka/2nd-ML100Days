```python
import os 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt


dir_data='./data/'

f_train=os.path.join(dir_data,'application_train.csv')
train=pd.read_csv(f_train)
train.head()
```

```python
train.corr()['TARGET'].sort_values()
```

```python
plt.plot( train.EXT_SOURCE_3,train.TARGET , '.')
plt.xlabel('EXT_SOURCE_3')
plt.ylabel('TARGET')
plt.show()
```

```python
mdata=train.groupby(['EXT_SOURCE_3'])['TARGET']
a=np.array(mdata)
plt.boxplot(a,labels='gg')
```
